```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("melbourne_housing.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
col = [
    "Suburb",
    "Rooms",
    "Type",
    "Method",
    "SellerG",
    "Regionname",
    "Propertycount",
    "Distance",
    "CouncilArea",
    "Bedroom2",
    "Car",
    "Landsize",
    "BuildingArea",
    "Price",
]
```

```python
df = df[col]
```

```python
df
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.isna().sum().sum()
```

```python
col_fill = ["Propertycount", "Distance", "Bedroom2", "Car"]
```

```python
df[col_fill] = df[col_fill].fillna(0)
```

```python
df
```

```python
df.isna().sum()
```

```python
df["Landsize"] = df["Landsize"].fillna(df["Landsize"].mean())
```

```python
df.isna().sum()
```

```python
df["BuildingArea"] = df["BuildingArea"].fillna(df["BuildingArea"].mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.isna().sum().sum()
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
x = df.drop(["Price"], axis=1)
```

```python
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=2
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.linear_model import Lasso
```

```python
model_lasso = Lasso(alpha=25, max_iter=500, tol=0.1)
```

```python
model_lasso.fit(x_train, y_train)
```

```python
model_lasso.score(x_train, y_train)
```

```python
model_lasso.score(x_test, y_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model_ridge = Ridge()
model_ridge.fit(x_train, y_train)
model_ridge.score(x_test, y_test)
```

There are four ways to get a balanced fit model.
K Fold Cross-Validation.
L1 and L2 Regularization.
Principal Component Analysis.
Bagging & Boosting.
