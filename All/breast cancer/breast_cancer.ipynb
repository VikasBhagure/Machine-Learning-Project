## Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("breast_cancer_train.csv")
```

```python
df
```

## Data Cleaning

```python
df = df.drop(df[df["bare_nuclei"]=="?"].index,axis=0)
```

```python
df["bare_nuclei"] = df["bare_nuclei"].astype(int)
```

```python
df.dtypes
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
df
```

```python
df.describe()
```

## Data Modelling

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=10)
model.fit(x,y)
```

```python
val_result = pd.read_csv('validation_results.csv')
```

```python
val_result.head()
```

```python
val = pd.read_csv('breast_cancer_validation.csv')
```

```python
val.head()
```

```python
y_pred = model.predict(val)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_val = val_result.iloc[:,-1]
```

```python
accuracy_score(y_val,y_pred)
```

```python
confusion_matrix(y_val,y_pred)
```

```python
result = pd.read_csv("breast_cancer_test.csv")
```

```python
result.head()
```

## For final apply for test csv

```python
predition = model.predict(result)
```

```python
data = {"id":result["id"],"class":predition}
```

```python
dataset = pd.DataFrame(data)
```

```python
dataset
```
